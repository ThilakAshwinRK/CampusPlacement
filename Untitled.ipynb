```python
def countSetBits(num):
    l=[0]*(num+1)  
    l[0]=0
    for i in range(1,num+1):
        if(i&1):
            l[i]=l[i-1]+1
        else:
            l[i]=l[i/2]
    print l
    return sum(l)
```

```python
print countSetBits(17)
```

```python
def numberOfBits(num):
    count=0
    while(num>0):
        if(num&1):
            count+=1
        num=num>>1
    return count
```

```python
print numberOfBits(6)
```

```python
def positionOfLeftmostSetBit(num):
    pos=0
    while(num>1):
        num=num>>1
        pos+=1
    return pos
```

```python
positionOfLeftmostSetBit(17)
```

```python
def minDif(a):
    a.sort()
    minDiff=1000
    for i in range(1,len(a)):
        if(minDiff>a[i]-a[i-1]):
            minDiff=a[i]-a[i-1]     
    return minDiff
```

```python
l=[30, 5, 20, 9]
print minDif(l)
```

```python
def gcd(a,b):
    if b==0:
        return a
    else:
        return gcd(b,a%b)
```

```python
gcd(2,7)
```
